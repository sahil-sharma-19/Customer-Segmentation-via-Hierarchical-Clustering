```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#hierarchical clustring : algorithm of unspervised learning groups the data into clusters
# No. of clusters is decided by dendrogram
# Dendrogram is tree like structure
```

```python
df=pd.read_csv("hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv")
df
```

```python
data=df.iloc[:,3:5].values
data
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customers Dendograms")
dend=shc.dendrogram(shc.linkage(data,method="ward"))
# The linkage() function is used to determine how clusters are merged in hierarchical clustering.
# It takes a dataset and computes a linkage matrix , which contains the distances between clusters
# Ward tries to keep the people in each group as similar as possible.
# It looks at all possible merges.
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=5,linkage="ward")
labels_=cluster.fit_predict(data) # fit_product trains the data point and assigns them the clusters
```

```python
# 5 clusters : 0,1,2,3,4
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data[:,0],data[:,1],c=cluster.labels_,cmap="rainbow")
```
